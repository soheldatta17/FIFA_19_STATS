```python
import pandas as pd

df = pd.read_csv('/content/Train.csv')

# Drop rows with NaN values only in critical columns
critical_columns = ['ID', 'Name', 'Age', 'Nationality', 'Club', 'Position', 'Overall']
df = df.dropna(subset=critical_columns)

# Alternatively, replace NaN values with a specific value for non-critical columns
df = df.fillna({
    'Photo': 'Unknown',
    'Flag': 'Unknown',
    'Club Logo': 'Unknown',
    'Value': 0,
    'Wage': 0,
    'Special': 0,
    'Preferred Foot': 'Unknown',
    'Body Type': 'Unknown',
    'Real Face': 'Unknown',
    'Jersey Number': 0,
    'Joined': 'Unknown',
    'Loaned From': 'Unknown',
    'Contract Valid Until': 'Unknown',
    'Height': 0,
    'Weight': 0,
    'Release Clause': 0
})

print(df.head())
```

```python
positions = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW',
             'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM',
             'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB',
             'RCB', 'RB']
```

```python
# Find the top 5 players for each position
top_players = {}
for pos in positions:
    pos_df = df[df['Position'] == pos]
    top_players[pos] = pos_df.sort_values(by='Potential', ascending=False).head(10)[['ID', 'Name', 'Age', 'Potential']].values.tolist()

# Print top players for each position in a readable format
for pos, players in top_players.items():
    print(f"Position: {pos}")
    print(f"{'ID':<10} {'Name':<25} {'Age':<5} {'Potential':<10}")
    print('-' * 50)
    for player in players:
        print(f"{player[0]:<10} {player[1]:<25} {player[2]:<5} {player[3]:<10}")
    print()
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv('/content/Train.csv')

# Define position categories
forwards_positions = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW']
midfielders_positions = ['LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 'LDM', 'CDM', 'RDM']
defenders_positions = ['LCB', 'CB', 'RCB']
left_back_positions = ['LB', 'LWB']
right_back_positions = ['RB', 'RWB']
goalkeeper_position = 'GK'

# Sort all players by 'Potential' in descending order
sorted_df = df.sort_values(by='Potential', ascending=False)

# Fetch top players for each category
top_forwards = sorted_df[sorted_df['Position'].isin(forwards_positions)].head(5)
top_midfielders = sorted_df[sorted_df['Position'].isin(midfielders_positions)].head(5)
top_defenders = sorted_df[sorted_df['Position'].isin(defenders_positions)].head(5)
top_left_backs = sorted_df[sorted_df['Position'].isin(left_back_positions)].head(1)
top_right_backs = sorted_df[sorted_df['Position'].isin(right_back_positions)].head(1)
top_goalkeeper = sorted_df[sorted_df['Position'] == goalkeeper_position].head(1)

# Combine all selected players into a single list
selected_players = {
    'Forward': top_forwards,
    'Midfielder': top_midfielders,
    'Defender': top_defenders,
    'Left Back': top_left_backs,
    'Right Back': top_right_backs,
    'Goalkeeper': top_goalkeeper
}

# Print the World XI lineup in a 4-3-3 formation
print("World XI in 4-3-3 Formation:")
print(f"{'No.':<5} {'Position':<15} {'Name':<20} {'Age':<1} {'Potential':<10}")
print('-' * 60)

# Initialize serial number
serial_no = 1

# Print Goalkeeper
if not selected_players['Goalkeeper'].empty:
    player = selected_players['Goalkeeper'].iloc[0]
    print(f"{serial_no:<5} Goalkeeper   {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Left Back (1 in the lineup)
for player in selected_players['Left Back'].itertuples():
    print(f"{serial_no:<5} Left Back    {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Right Back (1 in the lineup)
for player in selected_players['Right Back'].itertuples():
    print(f"{serial_no:<5} Right Back   {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Defenders (4 in a typical 4-3-3 formation)
defenders = selected_players['Defender'].head(4)  # Ensure there are 4 defenders
for player in defenders.itertuples():
    print(f"{serial_no:<5} Defender     {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Midfielders (3 in a typical 4-3-3 formation)
midfielders = selected_players['Midfielder'].head(3)  # Ensure there are 3 midfielders
for player in midfielders.itertuples():
    print(f"{serial_no:<5} Midfielder   {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Forwards (3 in a typical 4-3-3 formation)
forwards = selected_players['Forward'].head(3)  # Ensure there are 3 forwards
for player in forwards.itertuples():
    print(f"{serial_no:<5} Forward      {player.Name:<25} {player.Age:<3} {player.Potential:<10}")
    serial_no += 1

# Print Left Back (1 in a typical 4-3-3 formation)
```
